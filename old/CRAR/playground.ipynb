```python
import torch
import numpy as np
import seaborn as sns
```

```python
sz = 8
r = np.tile(np.arange(0, sz) / sz, (sz, 1)).T
```

```python
sns.heatmap(r, annot=True)
```

```python
theta = np.arange(0, sz) / sz * 2.0 * np.pi
theta = np.tile(theta, (sz, 1))
theta = np.roll(theta, 1, axis=1)
```

```python
sns.heatmap(theta, annot=True, cmap="YlGnBu")
```
